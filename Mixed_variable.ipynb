#### Here logic is that, first extrect the numeric value from the dataframe. If the dataframe has "Nan" value then old value is stored into the categorical column

```python
df["number_numerical"]=pd.to_numeric(df["number"],errors='coerce',downcast='integer')
```

```python
df.head()
```

```python
df['number_categorical']=np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
df["Cabin_numeric"]=df['Cabin'].str.extract('(\d+)')
df["Cabin_categorical"]=df["Cabin"].str[0]
```

```python
df.sample(5)
```

```python
df["Ticket"].unique()
```

```python
df["ticket_num"]=df["Ticket"].apply(lambda s: s.split()[-1])
df["ticket_num"]=pd.to_numeric(df["ticket_num"],errors='coerce',downcast='integer')
df["ticket_cat"]=df["Ticket"].apply(lambda s: s.split()[-1])
df["ticket_cat"]=np.where(df["ticket_cat"].str.isdigit(), np.nan, df["ticket_cat"])
```
